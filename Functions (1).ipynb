```python
def cake_baker():      #define a function
  print("Im ready to bake the cake")

cake_baker()      #invoke a function
```

```python
def cake_baker(p,f,s):      #define a function (p,f,s) are the parameters to be passed
  print("no of pieces =",p)
  print("flavour =",f)
  print("size =",p)

cake_baker("5","chocolate","2kg")      # insider brackets () they are arguments
cake_baker("2","vanilla","5kg")
```

```python
def add(*args):     #* will pack all those numbers in a tuple when you dont know how many parameters you will pass
  sum=0
  for i in args:
    sum+= i
  return sum
print(add(1,2,3,4,5))
```

```python
def add(*bunch):
  sum=0
  for i in bunch:
    sum+= i
  return sum
print(add(1,2,3,4,5))
```

```python
#kwargs i.e keyword arguments this will pass both keyword and arguments and here two asterisk will be used (**)
def greet(**kwargs):
  print("hello")
  for key,value in kwargs.items():
    print(value)

greet(name="tejaswini" , usn="01fe21bec035" , branch="enc")
```

```python
def greet(**kwargs):
  print("hello")
  for key,value in kwargs.items():
    print(value, end="  ")

greet(name="tejaswini" , usn="01fe21bec035" ,branch="enc")
```
