```python
#Simple if-else
year = int(input("Enter a year: "))

if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print(f"{year} is a leap year.")
else:
    print(f"{year} is not a leap year.")
```

```python
#string compare
stored_password = "kletech"

user_password = input("Enter your password: ")

if user_password == stored_password:
    print("Password is correct. Access granted.")
else:
    print("Incorrect password. Access denied.")
```

```python
#logical opration
price = float(input("Enter the price of the product: "))

if price > 1000:
    category = "High-end"
elif price >= 500 and price <= 1000:
    category = "Mid-range"
else:
    category = "Budget"

print(f"The product falls into the '{category}' category.")
```

```python
#Tuple data
title = input("Enter the title of the book: ")
author = input("Enter the author of the book: ")
genre = input("Enter the genre of the book: ")

book = (title, author, genre)

title = book[0]
author = book[1]
genre = book[2]

print(f"The book '{title}' by {author} is in the {genre} genre.")
```

```python
#simple else if
distance = float(input("Enter the distance of your journey in kilometers: "))

if distance <= 1:
    transportation_mode = "Walking"
elif distance <= 10:
    transportation_mode = "Biking"
elif distance <= 100:
    transportation_mode = "Driving"
else:
    transportation_mode = "Flying"

print(f"For a {distance} km journey, transportation is: {transportation_mode}")
```
